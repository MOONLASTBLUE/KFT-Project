```python
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
dataset = pd.read_csv('KFT Dataset.csv')
dataset['Flavor Tags'] = dataset['Flavor Tags'].fillna('').apply(lambda x: x.split(', '))
dataset
```

```python
# Train Word2Vec model
model = Word2Vec(sentences=dataset['Flavor Tags'], vector_size=50, window=5, min_count=1, workers=4)
```

```python
# Get embedding for a single tag or a list of tags
def get_tag_embedding(tags):
    tag_vectors = [model.wv[tag] for tag in tags if tag in model.wv]
    if tag_vectors:
        # Average vector for selected tags
        return sum(tag_vectors) / len(tag_vectors)  
    return None

# Find the most similar menu items based on selected tags
def recommend_using_word2vec(selected_tags, dataset):
    query_vector = get_tag_embedding(selected_tags)
    if query_vector is None:
        return []

    # Compute cosine similarity for all menu items
    similarities = []
    for _, row in dataset.iterrows():
        tags = row['Flavor Tags']
        item_vector = get_tag_embedding(tags)
        if item_vector is not None:
            sim = cosine_similarity([query_vector], [item_vector])[0][0]
            similarities.append(sim)
        else:
            similarities.append(0)

    dataset['Similarity'] = similarities
    return dataset.sort_values(by='Similarity', ascending=False).head(5)[['Menu', 'Similarity']]
```
